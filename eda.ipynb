```python
import pandas as pd

# Load Parquet file into a pandas DataFrame
df = pd.read_parquet("train-00000-of-00002.parquet", engine="pyarrow")


# Print the "persona" column of the first row
print(df.loc[17786, "persona"])
```

```python
%matplotlib inline
import re


# Ensure each persona is unique
df_unique = df.drop_duplicates(subset=["persona"]).copy()

# Function to parse persona strings into dictionaries
def parse_persona(persona_string):
    if not isinstance(persona_string, str) or not persona_string.strip():
        return None  # Ignore empty or non-string values
    
    # Regular expression to extract key-value pairs (handles multiline attributes)
    pattern = r"([\w\s\-]+):\s(.+)"
    matches = re.findall(pattern, persona_string)

    if matches:
        return {key.strip(): value.strip() for key, value in matches}  # Trim spaces
    return None

# Apply parsing function
df_unique["persona"] = df_unique["persona"].apply(parse_persona)

# Remove any rows where persona failed to parse
df_unique = df_unique.dropna(subset=["persona"])

# Convert list of dictionaries to a DataFrame
persona_df = pd.json_normalize(df_unique["persona"])

# Debugging: Check parsed data
print("\nParsed persona data (first 5 rows):")
print(persona_df.head())

# Select relevant attributes for visualization
categorical_columns = [
    "sex", "race", "ancestry", "household language", "education",
    "employment status", "class of worker", "industry category", "occupation category",
    "marital status", "household type", "citizenship", "veteran status",
    "disability", "health insurance", "political views", "religion"
]

numerical_columns = ["age"]

# Ensure figures are shown properly in different environments
plt.ion()

# Plot categorical distributions
for column in categorical_columns:
    if column in persona_df.columns and not persona_df[column].isna().all():
        print(f"ðŸ“Š Generating plot for {column}...")
        plt.figure(figsize=(10, 5))
        top_categories = persona_df[column].value_counts().nlargest(15)  
        sns.barplot(y=top_categories.index, x=top_categories.values, palette="viridis")
        plt.xlabel("Count")
        plt.ylabel(column)
        plt.title(f"Distribution of {column} (Top 15)")
        plt.show()

# Plot numerical distributions (e.g., age)
for column in numerical_columns:
    if column in persona_df.columns and not persona_df[column].isna().all():
        print(f"ðŸ“Š Generating plot for {column}...")
        plt.figure(figsize=(8, 5))
        sns.histplot(persona_df[column], bins=20, kde=True, color="blue")
        plt.xlabel(column)
        plt.ylabel("Count")
        plt.title(f"Distribution of {column}")
        plt.show()
```
